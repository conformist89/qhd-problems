```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

#  $Ð“_{Z \rightarrow f\overline{f} }  = \frac{|\overline{M_{fi}}|^2}{16\pi m_Z^2}(1-(\frac{2m_f}{m_Z})^2)^\frac{3}{2}\theta(m_Z - 2m_f)$

```python
class Particle:
    def __init__(self, name, mass, legend,gv, ga):
        self.name = name
        self.mass = mass
        self.legend = legend
        
        self.gv = gv
        self.ga = ga
```

```python
e = Particle('e', 0.0005, r'$H\rightarrow e^{+}e^{-}$', -0.038, 0.5)
tau = Particle('tau', 1.176, r'$H\rightarrow \tau^{+}\tau^{-}$', -0.038, 0.5)
muon = Particle('mu', 0.105, r'$H\rightarrow \mu^{+}\mu^{-}$', -0.038, 0.5)

t = Particle('t', 172, r'$H\rightarrow t^{+}t^{-}$', 0.192, 0.5)
b = Particle('b', 4.67, r'$H\rightarrow b^{+}b^{-}$', -0.346, -0.5)

c = Particle('c', 1.27, r'$H\rightarrow c^{+}c^{-}$', 0.192, 0.5)
d = Particle('d', 0.005, r'$H\rightarrow d^{+}d^{-}$', -0.346, 0.5)

s = Particle('s', 0.101, r'$H\rightarrow s^{+}s^{-}$', -0.346, 0.5)
u =Particle('u', 0.001, r'$H\rightarrow u^{+}u^{-}$', 0.192, -0.5)

```

```python
channels = [e, tau, t, b, c, d, s, u]
```

```python
# Z boson mass range
mz = np.linspace(0.00001,5,200)
```

```python
def amplitude(mz, ferm):
    return (ferm.gv**2)*(1+2*(ferm.mass**2) / (mz**2)) + (ferm.ga**2) * (1 - 4 * (ferm.mass**2) / (mz**2))
```

```python
def width(mz, ferm):
    term1 = mz - 2* ferm.mass
    term1[term1<0] = 0
    
    term2 = 1 - (2*ferm.mass / mz) ** 2
    term2[term2<0] = 0
    term2 = term2**(1/2)
    
    theta = np.ones(len(mz))
    
    return  amplitude(mz, ferm)**2 * term1 * term2
```

```python
widths_array = np.array([width(mz, f) for f in channels])
widths_sum = np.sum(widths_array, axis = 0)
```

```python
def plot_branching(mh, log_y_axis):
    
    
    fig, ax = plt.subplots(figsize=(20, 10))
    
    fontP = FontProperties()
    fontP.set_size('xx-large')
    
    for ferm in channels:
    
        plt.plot(mh, width(mh, ferm) / widths_sum, label = ferm.legend, linewidth=5)
    
    ax.xaxis.set_tick_params(labelsize=25)
    ax.yaxis.set_tick_params(labelsize=25)
    

    ax.set_xlabel(r'$m_H$[GeV] ', fontsize = 25)
    ax.set_ylabel('BR', fontsize = 25)
    
    if log_y_axis==1:
    
        ax.set_yscale('log')
    
    plt.legend(shadow=True, bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP, fontsize = 30)
    
    fig.tight_layout()
```

```python
plot_branching(mz, 0)
```

```python
plot_branching(mz, 1)
```
