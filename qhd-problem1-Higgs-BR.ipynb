```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

# $Ð“_{H \rightarrow f\overline{f} }  = \frac{m_f^2}{8\pi v^2} m_{H}(1-(\frac{2m_f}{m_H})^2)^\frac{3}{2}\theta(m_H - 2m_f)$

```python
class Particle:
    def __init__(self, name, mass, legend, n):
        self.name = name
        self.mass = mass
        self.legend = legend
        self.n = n
```

```python
#fermions mass

e = Particle('e', 0.0005, r'$H\rightarrow e^{+}e^{-}$',1)
tau = Particle('tau', 1.176, r'$H\rightarrow \tau^{+}\tau^{-}$',1)
muon = Particle('mu', 0.105, r'$H\rightarrow \mu^{+}\mu^{-}$',1)

t = Particle('t', 172, r'$H\rightarrow t^{+}t^{-}$',3)
b = Particle('b', 4.67, r'$H\rightarrow b^{+}b^{-}$',3)

c = Particle('c', 1.27, r'$H\rightarrow c^{+}c^{-}$',3)
d = Particle('d', 0.005, r'$H\rightarrow d^{+}d^{-}$',3)

s = Particle('s', 0.101, r'$H\rightarrow s^{+}s^{-}$',3)
u =Particle('u', 0.001, r'$H\rightarrow u^{+}u^{-}$',3)
```

```python
channels = [e, tau, t, b, c, d, s, u]
```

```python
# Higgs boson mass range
mh = np.linspace(0.00001,5,200)
```

```python
theta = np.ones(len(mh))
```

```python
def width(mh, ferm):
    term1 = mh - 2* ferm.mass
    term1[term1<0] = 0
    
    term2 = 1 - (2*ferm.mass / mh) ** 2
    term2[term2<0] = 0
    term2 = term2**(3/2)
    
    theta = np.ones(len(mh))
    
    return (ferm.mass**2) * mh * term1 * theta * term2 * ferm.n
```

```python
widths_array = np.array([width(mh, f) for f in channels])
```

```python
widths_sum = np.sum(widths_array, axis = 0)
```

```python
def plot_branching(mh, log_y_axis):
    
    
    fig, ax = plt.subplots(figsize=(20, 10))
    
    fontP = FontProperties()
    fontP.set_size('xx-large')
    
    for ferm in channels:
    
        plt.plot(mh, width(mh, ferm) / widths_sum, label = ferm.legend, linewidth=5)
    
    ax.xaxis.set_tick_params(labelsize=25)
    ax.yaxis.set_tick_params(labelsize=25)
    

    ax.set_xlabel(r'$m_H$[GeV] ', fontsize = 25)
    ax.set_ylabel('BR', fontsize = 25)
    
    if log_y_axis==1:
    
        ax.set_yscale('log')
    
    plt.legend(shadow=True, bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP, fontsize = 30)
    
    fig.tight_layout()
```

```python
plot_branching(mh, 0)
```

```python
plot_branching(mh, 1)
```
